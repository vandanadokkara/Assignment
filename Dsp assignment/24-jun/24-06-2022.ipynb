```python
import matplotlib.pyplot as plt
names=["vandana","manoj","joshna","kamala","apparao"]
salary=[15000,20000,25000,10000,23000]
plt.bar(names,salary,color="brown")
plt.title("Details in home",color="Red")
plt.xlabel("name")
plt.ylabel("salary")
plt.show()
```

```python
#Horizental Bars
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
# Bar color
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="red")
plt.show()
```

```python
#Bar width
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width=0.1)
plt.show()
```

```python
#Bar Height
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,height=0.8)
plt.show()
```
