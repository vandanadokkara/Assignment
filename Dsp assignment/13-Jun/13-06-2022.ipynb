```python
# Read total profit of all months and show it using a lineplot
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('vandana.csv')
profitlist=data['total_profit'].tolist()
monthlist=data['month'].tolist()
plt.plot(monthlist,profitlist,label='month-wise profit data of last year')
plt.xlabel('month number')
plt.ylabel('profit in dollar')
plt.title('Company profit per month')
plt.show()
```

```python
# For this line showing with some style propertices
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('vandana.csv')
profit=data['total_profit'].tolist()
month=data['month'].tolist()
plt.plot(month,profit,label='month-wise profit data in lastyear',marker='*',linestyle='--',mec='r',color='navy',linewidth='5')
plt.xlabel('month number')
plt.ylabel('profitlist')
plt.title('Company profit per month')
plt.show()
```

```python
# Read all product sales data and show it using a multiline plot
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('vandana.csv')
monthlist=df['month'].tolist()
faceCreamSalesData=df['facecream'].tolist()
faceWashSalesData=df['facewash'].tolist()
toothPasteSalesData=df['toothpaste'].tolist()
bathingsoapsalesDate=df['bathingsoap'].tolist()
shampooSalesDate=df['shampoo'].tolist()
moisturizerSalesDate=df['moisturizer'].tolist()
plt.plot(monthlist,faceCreamSalesData,label='face cream sales date',marker='o',color='hotpink',linewidth=3)
plt.plot(monthlist,faceWashSalesData,label='face wash sales date',marker='o',color='red',linewidth=3)
plt.plot(monthlist,toothPasteSalesData,label='tooth paste sales data',marker='o',color='navy',linewidth=3)
plt.plot(monthlist,bathingsoapsalesDate,label='bathing soap sales date',marker='o',color='black',linewidth=3)
plt.xlabel('month number')
plt.ylabel('sales units in number')
plt.legend(loc='upper left')
plt.title('sales date')
plt.show()
```
