```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Profit.csv")
print(data)
profitlist=data['Profit'].tolist()
monthlist=data['Month'].tolist()
plt.plot(monthlist,profitlist,label='month-wise profit data of last year')
plt.xlabel('month number')
plt.ylabel('profit in dollar')
plt.title('company profit per month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Profit.csv")
print(data)
profitlist=data['Profit'].tolist()
monthlist=data['Month'].tolist()
plt.plot(monthlist,profitlist,label='month-wise profit data of last year',color='olive',marker="*",mec="red",linewidth='7')
plt.xlabel('month number')
plt.ylabel('profit in dollar')
plt.title('company profit per month')
plt.show()
```
