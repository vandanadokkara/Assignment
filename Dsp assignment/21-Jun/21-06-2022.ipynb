```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([1,50])
ypoints=np.array([1,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([5,200],[2,200],".-.")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
#multiple points
xpoints=np.array([1,2,3,4,5])
ypoints=np.array([3,6,4,9,7])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#without y-axis
xpoints=np.array([3,6,8,9])
plt.plot(xpoints)
plt.show()
```
