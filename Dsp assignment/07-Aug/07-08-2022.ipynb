```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=pd.Series([12,2,3,8,5,8,6])
mean=x.mean()
median=x.median()
sd=x.std()
varience=x.var()
mode=x.mode()
x=[1,2,3,4,5]
y=np.array([mean,median,sd,varience,mode],dtype=object)
labels=("m","me","sd","var","mod")
plt.bar(x,y,label=labels,color="red")
plt.legend(loc="upper right")
plt.show()
```
