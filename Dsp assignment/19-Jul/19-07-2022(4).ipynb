```python
#Creating a piechart
import matplotlib.pyplot as plt
import numpy as np
a=np.array([15,16,22,28,10,9])
plt.pie(a)
plt.show()
```

```python
#Adding labels
import matplotlib.pyplot as plt
import numpy as np
a=np.array([15,16,22,28,10,9])
my_label=["vandana","joshna","manoj","apparao","kamala","raju"]
plt.pie(a,labels=my_label)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([15,16,22,28,10,9])
my_label=["vandana","joshna","manoj","apparao","kamala","raju"]
my_colors=["black","blue","hotpink","green","red","olive"]
plt.pie(a,labels=my_label,colors=my_colors)
plt.show()
```

```python
#Creating a Bar graph
import matplotlib.pyplot as plt
import numpy as np
name=["vandana","joshna","manoj","apparao","kamala"]
salary=["100","400","200","150","500"]
plt.bar(name,salary)
plt.xlabel("name")
plt.ylabel("salary")
plt.show()
```
