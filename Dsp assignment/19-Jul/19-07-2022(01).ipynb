```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([5,15,25,35,45,55])
print(x)
x=np.array([5,15,25,35,45,55]).reshape(-1,1)
print(x)
y=np.array([5,20,14,32,22,38])
print(y)
plt.scatter(x,y)
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
print(model.intercept_)
print(model.coef_)
print(model.score(x,y))
```

```python
y_pred=model.predict(x)
print(y_pred)
print(model.score(x,y_pred))
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
print(model.intercept_+model.coef_*x)
```

```python
x_new=np.arange(5).reshape((-1,1))
print(x_new)
```

```python
y_new=model.predict(x_new)
print(y_new)
print(model.score(x_new,y_new))
```

```python
plt.scatter(x_new,y_new)
plt.plot(x_new,y_new)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data=pd.read_csv("vandu.csv")
plt.scatter(data['B'],data['Salary'],c="olive")
plt.xlabel('B')
plt.ylabel('Salary')
plt.show()
x=data['B'].values.reshape(-1,1)
y=data["Salary"].values
```

```python
reg=LinearRegression()
reg.fit(x,y)
print(reg.score(x,y))
print(reg.coef_)
print(reg.intercept_)
y_pred=reg.predict(x)
plt.scatter(data["B"],data["Salary"],c='green')
plt.plot(data["B"],y_pred,c='hotpink',linewidth=2)
plt.xlabel("B")
plt.ylabel("Salary")
plt.show()
print(reg.score(x,y_pred))
```
