```python
#startangle
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["vandana","joshna","manoj","raju"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
#Explode
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["vandana","joshna","manoj","raju"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
#shadow
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["vandana","joshna","manoj","raju"]
myexplode=[0.4,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
#Colors
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["vandana","joshna","manoj","raju"]
mycolors=["black","olive","c","navy"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
#legend
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["vandana","joshna","manoj","raju"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```
